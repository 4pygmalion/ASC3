```python
import mlflow
from revision_utils import TRACKING_URI, RUN_ASC3_W_RANKNET, RUN_ASC3_WO_RANKNET, FOLD_RESULTS_PICKLE

mlflow.set_tracking_uri(TRACKING_URI)

result_w_ranknet_path = mlflow.artifacts.download_artifacts(FOLD_RESULTS_PICKLE.format(run_id=RUN_ASC3_W_RANKNET))
result_wo_ranknet_path = mlflow.artifacts.download_artifacts(FOLD_RESULTS_PICKLE.format(run_id=RUN_ASC3_WO_RANKNET))
```

```python
import os
import sys
from revision_utils import open_pickle, DATA_PATH

REV_DIR = os.getcwd()
RES_DIR = os.path.dirname(REV_DIR)
ROOT_DIR = os.path.dirname(RES_DIR)
sys.path.append(ROOT_DIR)

result_w_ranknet = open_pickle(result_w_ranknet_path)
result_wo_ranknet = open_pickle(result_wo_ranknet_path)
patient_dataset = open_pickle(DATA_PATH)
```

```python
import os
import sys

import pandas as pd
revision_dir = os.getcwd()
result_dir = os.path.dirname(revision_dir)
root_dir = os.path.dirname(result_dir)

sys.path.append(root_dir)
from core.metric import Metric

rows = list()
method_names = ["3ASC W/ RankNet", "3ASC W/O RankNet"]
for method, method_result in zip(method_names, [result_w_ranknet, result_wo_ranknet]):
    for fold in range(0, 5):
        for k in [1, 2, 3, 4, 5, 10, 15, 20, 100]:
            snv_cnv_hit = Metric.topk_recall(
                bag_labels=method_result["fold_bag_y_trues"][fold],
                instance_labels=method_result["fold_instance_y_trues"][fold],
                instance_probs=method_result["fold_instance_y_probs"][fold],
                k=k,
                is_any=True
            )
            rows.append([method, fold, k, snv_cnv_hit])
            
snv_data = pd.DataFrame(rows, columns=["loss", "fold", "K", "HitRate@K\n(SNV+CNV)"])
snv_data.head()
```

```python
from core.data_model import PatientDataSet
from results.revision.revision_utils import get_cnv_only_case, get_topk_folds

ks = [1, 2, 3, 4, 5, 10, 15, 20, 100]
dfs = list()
for method, method_result in zip(method_names, [result_w_ranknet, result_wo_ranknet]):
    ranknet_res = {"fold_instance_y_trues": list(), "fold_instance_y_probs":list()}
    all_fold_test_ids = method_result["test_ids"]
    for fold in range(5):
        test_ids = all_fold_test_ids[fold]
        cnv_test_ids = get_cnv_only_case(patient_dataset[test_ids])
        test_indices = [test_ids.index(sample_id) for sample_id in cnv_test_ids]
        instance_y_trues = [method_result["fold_instance_y_trues"][fold][_id] for _id in test_indices]
        instance_y_probs = [method_result["fold_instance_y_probs"][fold][_id] for _id in test_indices]
        ranknet_res["fold_instance_y_trues"].append(instance_y_trues)
        ranknet_res["fold_instance_y_probs"].append(instance_y_probs)

    peformance = get_topk_folds(ranknet_res["fold_instance_y_trues"], ranknet_res["fold_instance_y_probs"])
    df = pd.DataFrame(peformance, columns=ks)
    df["loss"] = method
    dfs.append(df)
    
cnv_data = pd.concat(dfs)
cnv_data.index.name = "fold"
cnv_data.reset_index(inplace=True)
cnv_data = cnv_data.melt(id_vars=['fold', 'loss'], var_name='K', value_name='HitRate@K\n(CNV)')
cnv_data
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import mannwhitneyu
from statannot import add_stat_annotation

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
sns.boxplot(
    data=snv_data,
    x="K",
    y="HitRate@K\n(SNV+CNV)",
    hue="loss", 
    ax=axes[0]
)
pairs = [
    # (("3ASC W/ RankNet", k), ("3ASC W/O RankNet", k))
    ((k, "3ASC W/ RankNet"), (k, "3ASC W/O RankNet"))
    for k in [1, 2, 3, 4, 5, 10, 15, 20, 100]
]
p_valeus = [
    mannwhitneyu(
        snv_data.loc[(snv_data["K"]==k) & (snv_data["loss"]=="3ASC W/ RankNet")]["HitRate@K\n(SNV+CNV)"],
        snv_data.loc[(snv_data["K"]==k) & (snv_data["loss"]=="3ASC W/O RankNet")]["HitRate@K\n(SNV+CNV)"],
    ).pvalue
    for k in [1, 2, 3, 4, 5, 10, 15, 20, 100]
]
add_stat_annotation(
    axes[0],
    data=snv_data,
    x="K",
    y="HitRate@K\n(SNV+CNV)",
    hue="loss",
    box_pairs=pairs,
    perform_stat_test=False,
    text_format='star',
    loc='inside',
    verbose=0,
    comparisons_correction=None,
    pvalues=p_valeus,
)

sns.boxplot(
    data=cnv_data,
    x="K",
    y="HitRate@K\n(CNV)",
    hue="loss", 
    ax=axes[1]
)
pairs = [
    # (("3ASC W/ RankNet", k), ("3ASC W/O RankNet", k))
    ((k, "3ASC W/ RankNet"), (k, "3ASC W/O RankNet"))
    for k in [1, 2, 3, 4, 5, 10, 15, 20, 100]
]
p_valeus = [
    mannwhitneyu(
        cnv_data.loc[(cnv_data["K"]==k) & (cnv_data["loss"]=="3ASC W/ RankNet")]["HitRate@K\n(CNV)"],
        cnv_data.loc[(cnv_data["K"]==k) & (cnv_data["loss"]=="3ASC W/O RankNet")]["HitRate@K\n(CNV)"],
    ).pvalue
    for k in [1, 2, 3, 4, 5, 10, 15, 20, 100]
]
add_stat_annotation(
    axes[1],
    data=cnv_data,
    x="K",
    y="HitRate@K\n(CNV)",
    hue="loss",
    box_pairs=pairs,
    perform_stat_test=False,
    text_format='star',
    loc='inside',
    verbose=0,
    comparisons_correction=None,
    pvalues=p_valeus,
)

plt.tight_layout()
plt.savefig("results/3_5_ablation_test.png")
```
